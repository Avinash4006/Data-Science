```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_s_age=np.random.randint(18,45,(100))
py_s_age=np.random.randint(15,45,(100))
```

```python
print(ml_s_age)
print(py_s_age)
```

```python
plt.hist(ml_s_age,bins=None)
plt.title("Ml Stud. Age ")
plt.xlabel("Age Cat")
plt.ylabel("No.Of s") 
```

```python
plt.hist(py_s_age)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_s_age,bins,rwidth=0.8)
plt.title("Ml Stud. Age ")
plt.xlabel("Age Cat")
plt.ylabel("No.Of s") 
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_s_age,bins,rwidth=0.8,histtype='bar',
         color='g',label="Ml stu")
plt.hist(ml_s_age,bins,rwidth=0.8,histtype='bar',
         color='y',label="Py Student")
plt.title("Ml Stud. Age ")
plt.xlabel("Age Cat")
plt.ylabel("No.Of s") 
plt.legend()
```
