```python
df1=pd.DataFrame({'ID':[1,2,3,4],'class':[9,10,11,12]})
```

```python
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2,on="ID")
```

```python
pd.merge(df2,df1,on="ID")
```

```python
df3=pd.DataFrame({'ID':[1,2,3,5],'Name':['A','B','C','D']})
df3
```

```python
pd.merge(df1,df3,on='ID')
```

```python
pd.merge(df1,df3,on='ID',how='inner')
```

```python
pd.merge(df1,df3,on='ID',how='left')
```

```python
pd.merge(df1,df3,on='ID',how='right')
```

```python
pd.merge(df1,df3,on='ID',how='outer')
```

```python
pd.merge(df1,df3,on='ID',how='outer',indicator=True)
```
